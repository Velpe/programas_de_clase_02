cuando se tienen dos número que son muy parecidos entre si, se recomienda utilizar variables que tengan una buena precisión. CUando se resta, se cancelan las cifras menos significativas. Un pequeño error de poco porcentaje puede implicar un error considerable. Lo que sucede con la cancelación sustractiva es que los números menos significativos se pierden.

```python
#Cancelación sustractiva
```

```python
a=1.0
b=1.0
x_list=[]
y_list=[]
for i in range(1,10):
    c=10**(-i)
    x_1 = (-b+(b**2-4*a*c)**0.5)/(2*a)
    x_1_p=2*c/(-b-(b**2-4*a*c)**0.5)
    error=abs(x_1_p - x_1)/abs(x_1_p)*100
    x_list.append(c)
    y_list.append(error)
    print("c=%.5e x_1=%.16e x_1_p=%.16e error=%.16f"%(c,x_1,x_1_p,error))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(15,10))
ax=plt.axes()
ax.plot(x_list,y_list,"-.")
ax.grid()
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel("c")
ax.set_ylabel("error")
plt.show()
```
