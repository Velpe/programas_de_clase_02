para los número de punto flotante se utiliza el formato
S= signo, 1= bit fantasma, M (Mantisa), 2^e

la normalización es poner la coma del número decimal antes del primer bit significativo.

Para el reescalamiento de los exponentes negativos, se debe tomar un nuevo exponente al cuál se le debe sumar el bias que equivale a 1023

es decir, si el exponente es 3 se le debe sumar 1023=1026=10000000010

el bias es (2^n-1)-1, donde n es la cantidad de casillas para el exponente
